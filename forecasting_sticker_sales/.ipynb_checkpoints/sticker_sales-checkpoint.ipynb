In this project, I will try method 2 and 3. But first, let us try with one specific situations here.

```python
canada_premium_kaggle = df[df["country"] == "Canada"]
canada_premium_kaggle = canada_premium_kaggle[canada_premium_kaggle["store"] == "Premium Sticker Mart"]
canada_premium_kaggle = canada_premium_kaggle[canada_premium_kaggle["product"] == "Kaggle"]
canada_premium_kaggle
```

The series looks like it has some pattern every year.

```python
plt.figure(figsize=(20, 5))
plt.plot(canada_premium_kaggle["num_sold"])
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(canada_premium_kaggle["num_sold"])
print(f"ADF Statistic: {adf_test[0]}")
print(f"P-value: {adf_test[1]}")
```

`p-value` is larger than 0.05, so the time series is **not stationary**. We will try to differentiate it to make it stationary.

```python
canada_premium_kaggle["num_sold_diff"] = canada_premium_kaggle["num_sold"].diff()
canada_premium_kaggle.dropna(inplace=True)

plt.figure(figsize=(20, 5))
plt.plot(canada_premium_kaggle["num_sold_diff"])
plt.show()
```

```python
adf_test = adfuller(canada_premium_kaggle["num_sold_diff"])
print(f"ADF Statistic: {adf_test[0]}")
print(f"P-value: {adf_test[1]}")
```

```python
canada_premium_kaggle.count()
```

The series is **stationary** now. Let's try fitting a SARIMA model.

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(canada_premium_kaggle["num_sold"], order=(1, 0, 1), seasonal_order=(1, 0, 1, 7))

model_fit = model.fit()
model_fit.summary()
```

```python
# Forecast next 30 days
forecast = model_fit.get_forecast(steps=30)

# Create a date index starting from the last date in the dataset
forecast_index = pd.date_range(start=df.index[-1], periods=30, freq="D")

# Convert forecast predictions to a pandas Series
forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)

# Plot actual vs. forecast
plt.figure(figsize=(15, 5))
# plt.plot(diff_can, label="Observed", color="blue")
plt.plot(forecast_series, label="Forecast", color="red", linestyle="dashed")
plt.legend()
plt.title("Time Series Forecast")
plt.show()
```
